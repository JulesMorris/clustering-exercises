```python
df = df.sort_values(by = 'Healthy life expectancy', ascending = False)
df
```

```python
#create a boolean mask

df['Score'] >= 7
```

```python
#assign to variable
greater_than_seven = df['Score'] >= 7
```

```python
#pass the variable back into df to get subset data
df[greater_than_seven]
```

```python
#create boolean mask for multiple conditions

score_over_seven_life_expectancy_less_than_one = (df['Score'] >= 7 ) & (df['Healthy life expectancy'] < 1)
```

```python
#pass back into df
df[score_over_seven_life_expectancy_less_than_one]
```

```python
#pass conditions into subset directly for same result

df[(df['Score'] >= 7 ) & (df['Healthy life expectancy'] < 1)]

```

```python
df["GenerosityDP"] = df["GDP per capita"] * df["Generosity"]
```

```python
df.head()
```

```python
#create function that takes in score and assigns it to a category
def map_score_to_category(score):
    if score >= 7:
        return "High Score"
    elif score < 4:
        return "Low Score"
    else:
        return("Medium Score")
```

```python
#pass the function directly into the map method
df['Score'].map(map_score_to_category)
```

```python
#look at value counts
df['Score'].map(map_score_to_category).value_counts()
```

```python
#create new column
df['score_category'] = df['Score'].map(map_score_to_category)
df.head()
```

```python
#lambda creates boolean mask for countries starting with letter s

df["Country or region"].map(lambda x: x[0] == "S")
```

```python
#pass into df to return countries that start with S
df[df["Country or region"].map(lambda x: x[0] == "S")]
```

```python
#round scores
df["Score"].round()
```

```python
#get value counts for rounded scores
df["Score"].round().value_counts()
```

```python
#create new column for rounded scores
df["score_rounded"] = df["Score"].round()
```

```python
#use groupby on new column, the first part of groupby takes the data and splits it into groups
df.groupby("score_rounded")
```

```python
#assign to variable
gb = df.groupby("score_rounded")
```

```python
#look at the groups, a dictionary where key is rounded score, and value for each key is the 
# list of indices where score occurs
gb.groups
```

```python
#call aggregation method on grouped data

gb.mean()
```

```python
#look at what the gdp per capital value for each rounded score is
gb.mean()["GDP per capita"]
```
